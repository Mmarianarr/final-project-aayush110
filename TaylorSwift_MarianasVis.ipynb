```python
#Figure A. 
import altair as alt


dom = ['Taylor Swift','Lil Wayne','50 Cent', 'Adele', 'Backstreet Boys', 'Britney Spears', 
        'Eminem', 'Garth Brooks','Lady Gaga' ,'Limp Bizkit',
       'Norah Jones','NSYNC', 'The Beatles', 'Usher','Whitney Houston']

    
rng = ['red', 'pink','pink','pink','pink','pink',
       'pink','pink','pink','pink','pink','pink','pink','pink','pink']

alt.Chart(data).mark_bar().encode(
    #y="Album", 
    y=alt.Y("Album:N", sort=alt.SortField("Year", order="descending")),
    x ="Units(Millions)",
   # color = "Artist"
    color=alt.Color('Artist', scale=alt.Scale(domain=dom, range=rng)) 

)

# order the album by year 
```

```python
#Figure B.

import altair as alt
import pandas as pd

# Load the data
spotify = pd.read_csv("spotify_data.csv")


artist_count = spotify.sort_values(by="Total number of streams", ascending=False)[10:350]


dropdown = alt.binding_select(options = artist_count["Artist"].unique(), name = "Select a artist:")

selection = alt.selection(type = 'single', fields = ['Artist'], bind = dropdown)


alt.Chart(artist_count).mark_circle().encode(
    
    y = alt.Y("mean(Total Number of Songs)", axis=alt.Axis(title="Total Number of Songs")),
    
    x= "Total number of streams", 
  
    color= alt.Color("Artist", scale= alt.Scale(scheme = 'spectral')),   
    size = "Total number of streams",    
    tooltip=["Artist", "Total Number of Songs", "Artist Followers2"],    
    opacity = alt.condition(selection, alt.value(1), alt.value(0.15))
    
).add_selection(selection)


#c1|c2


```
